```python
import numpy as np
import matplotlib.pyplot as plt
import math

#our aim is to find the t which makes e^(-R0*t)+t-1 = 0
```

```python
# e^(-R0*t)+t-1
def f(t, R0):
    return math.e**(-R0*t)+t-1
f = np.vectorize(f)

def draw_for_R0(R0):
    x = np.linspace(0, 1, 10000)
    y = f(x, R0)

    plt.figure(figsize=(15,5))
    plt.plot(x, y)
    plt.title("R0 = "+(R0.__str__()))
    plt.xlabel("t")
    plt.show()

draw_for_R0(1.1)
draw_for_R0(1.5)
draw_for_R0(1.9)
```

```python
# We can see that f(0)=0, f(0+epsilon) < 0 for R0>=1, f(1) > 0.
# So we can apply bijection method between epsilon and 1

def solve_for_R0(R0):
    start = 1e-5 #epsilon
    end   = 1
    while end-start>1e-6:
        mid = (start+end)/2
        if f(mid, R0) < 0:
            start = mid
        else:
            end = mid
    return start

solve_for_R0 = np.vectorize(solve_for_R0)

x = np.linspace(1, 5, 1000)
y = solve_for_R0(x)

plt.figure(figsize=(15,5))
plt.plot(x, y)
plt.xlabel("R0")
plt.ylabel("Final Size t")
plt.show()
```
